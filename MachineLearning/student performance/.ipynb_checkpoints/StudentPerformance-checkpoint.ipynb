```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to all categorical columns
train['test preparation'] = label_encoder.fit_transform(train['test preparation'])
train['gender'] = label_encoder.fit_transform(train['gender'])
train['parental level of education'] = label_encoder.fit_transform(train['parental level of education'])
train['lunch'] = label_encoder.fit_transform(train['lunch'])
train['Section'] = label_encoder.fit_transform(train['Section'])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import pandas as pd

# Apply One-Hot Encoding
train_encoded = pd.get_dummies(train, columns=['test preparation', 'gender', 'parental level of education', 'lunch', 'Section'])

# Define features and target
X = train_encoded.drop(['Roll no', 'exam score'], axis=1)
y = train_encoded['exam score']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and fit the model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Get feature importances
importances = model.feature_importances_
features = X.columns

# Create a DataFrame for feature importances
importance_df = pd.DataFrame({'Feature': features, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print(importance_df)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=importance_df)
plt.title('Feature Importances from Random Forest')
plt.show()
```

```python
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor

# Define your model parameters dictionary
model_params = {
    'random_forest_regressor': {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': [50, 100],
            'max_depth': [10, 20],
            'min_samples_split': [2, 5]
        }
    },
    'svm': {
        'model': SVR(),
        'params': {
            'kernel': ['linear', 'rbf'],
            'C': [1, 10],
            'epsilon': [0.1, 0.2]
        }
    },
    'gradient_boosting_regressor': {
        'model': GradientBoostingRegressor(),
        'params': {
            'learning_rate': [0.1, 0.2],
            'n_estimators': [50, 100],
            'max_depth': [3, 5]
        }
    } #,
    # 'mlp_regressor': {
    #     'model': MLPRegressor(max_iter=500),
    #     'params': {
    #         'hidden_layer_sizes': [(50,), (100,)],
    #         'activation': ['relu', 'tanh'],
    #         'solver': ['adam'],
    #         'alpha': [0.0001],
    #         'learning_rate': ['constant']
    #     }
    # }
}
```

```python
# Initialize the k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=56)

# List to store the results
scores = []

# Iterate over each model in model_params
for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=kf, return_train_score=False, error_score='raise')
    clf.fit(X_train, y_train)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

# Create a DataFrame from the scores list
best_model = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

# Print or return the best models and their parameters
print(best_model)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# After running the GridSearchCV and storing results in `best_model` DataFrame
# Find the best model
best_model_info = best_model.loc[best_model['best_score'].idxmax()]
best_model_name = best_model_info['model']
best_params = best_model_info['best_params']

# Initialize the best model with the best parameters
best_model_class = model_params[best_model_name]['model']
best_model_instance = best_model_class.set_params(**best_params)

# Fit the best model on the training data
best_model_instance.fit(X_train, y_train)

# Predict on the test data
y_pred = best_model_instance.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Best Model: {best_model_name}")
print(f"Best Parameters: {best_params}")
print(f"Mean Squared Error on Test Data: {mse:.2f}")
print(f"R-squared Score on Test Data: {r2:.2f}")
```
