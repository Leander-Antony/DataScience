```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("imyjoshua/average-time-spent-by-a-user-on-social-media")

print("Path to dataset files:", path)
```

```python
import pandas as pd
df = pd.read_csv(path + "/dummy_data.csv")
df.head()
```

```python
from datetime import time
import plotly.express as px

fig = px.bar(df, x='platform', y='time_spent', color='gender', labels={'time_spent':'time spent (hours)'}, barmode='group')
fig.show()
```

```python
# Pie chart for gender distribution
fig = px.pie(df, names='gender', title="Gender Distribution of Social Media Users", labels={'gender': 'Gender'})
fig.show()
```

```python
# Box plot for age by platform
fig = px.box(df, x='platform', y='age', color='platform', title="Age Distribution by Platform")
fig.show()
```

```python
# Scatter plot for income vs time spent
fig = px.scatter(df, x='income', y='time_spent', color='platform',
                 size='age', title="Income vs. Time Spent on Social Media",
                 labels={'income':'Income (USD)', 'time_spent':'Time Spent (hours)'})
fig.show()
```

```python
# Bar plot for interests across platforms
fig = px.bar(df, x='interests', y='time_spent', color='platform',
             title="Interests Across Platforms", barmode='group')
fig.show()
```

```python
# Histogram for age distribution
fig = px.histogram(df, x='age', nbins=5, title="Age Distribution of Social Media Users", labels={'age':'Age'})
fig.show()
```

```python

import plotly.graph_objects as go

# Create a correlation matrix
corr_matrix = df[['age', 'time_spent', 'income']].corr()

# Heatmap for the correlation matrix
fig = go.Figure(data=go.Heatmap(z=corr_matrix.values,
                                x=corr_matrix.columns,
                                y=corr_matrix.columns))
fig.update_layout(title="Correlation Heatmap of Numerical Variables")
fig.show()
```

```python
# Bar plot for home ownership and car ownership
df_ownership = df[['isHomeOwner', 'Owns_Car']].melt(var_name='Ownership Type', value_name='Ownership Status')
ownership_counts = df_ownership.groupby(['Ownership Type', 'Ownership Status']).size().reset_index(name='Count')

fig = px.bar(ownership_counts, x='Ownership Type', y='Count', color='Ownership Status',
             title="Home and Car Ownership of Social Media Users", barmode='group')
fig.show()
```
