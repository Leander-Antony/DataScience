```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data from a CSV file hosted online
df = pd.read_csv("https://raw.githubusercontent.com/Leander-Antony/DataScience/main/data-viz/EDA%20world%20population/World%20Population%20Growth.csv")
df.head() # Displaying the first few rows
```

```python
df.tail() # Displaying the last few rows
```

```python
df.info() # Displaying information about the DataFrame structure
```

## 1.Trends over time:

How has the world population % changed over the years?

Are there any noticeable patterns or fluctuations in population growth rates?

```python
# Remove the '%' symbol from the "Yearly Growth %" column
df['Yearly Growth %'] = df['Yearly Growth %'].str.rstrip('%').astype(float)
sns.lineplot(data=df,x="Year",y="Yearly Growth %")
plt.title('world population % changed over the years')
plt.xlabel('Year')
plt.ylabel('Yearly Growth in percentage')
plt.show()
```

The reason for the decrease in '*Yearly Growth %*' in 1960 could due to the great chinese famine.

The reason for the decrease in '*Yearly Growth %*' in 2020 is due to the pandemic.

## 2. Yearly growth percentage:

What is the average yearly growth percentage of the world population?

```python
mean_growth = np.mean(df['Yearly Growth %'])
mean_growth_rounded = round(mean_growth, 3)

print("Average Yearly Growth Percentage:", mean_growth_rounded)
```

## 3. Population distribution:
How is the population distributed across different years?

Are there any years with particularly high or low population growth?

```python
plt.figure(figsize=(10, 8))
plt.plot(df['Year'], df['Population'], marker='o',color="blue")
plt.title('Population Distribution Across Different Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks(fontsize=6)
plt.grid(True, which='major')
plt.show()
```

There is no particular high or low population growth in any year

## 4.Number and density:
How does the total population number change over time?

Is there any correlation between population density and population growth?

```python
# Scatter plot for population density versus population growth
plt.figure(figsize=(8, 6))
plt.scatter(df['Density (Pop/km2)'], df['Yearly Growth %'], color='green', alpha=0.5)
plt.title('Population Density vs Population Growth')
plt.xlabel('Population Density')
plt.ylabel('Population Growth')
plt.grid(True)
plt.show()
```

we can clearly see that even if the '*Yearly Growth %*' decreases the '*Population Density*' is increasing the reason could be:

Urbanization and migration. Urban areas attract people seeking better opportunities and amenities, leading to concentrated growth despite a slowing overall rate. Moreover, advancements in infrastructure and healthcare extend life expectancy, sustaining or augmenting population density.
